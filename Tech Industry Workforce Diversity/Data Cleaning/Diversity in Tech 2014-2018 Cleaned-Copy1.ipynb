```python
import pandas as pd
from sqlalchemy import create_engine
import pymysql
pymysql.install_as_MySQLdb()
import numpy as np
```

```python
# import data
csv_file ="../Data Used & Resources/Employee Diversity in Tech 2.csv"
Tech_Data = pd.read_csv(csv_file)
Tech_Data.head()
```

```python
# Data Cleaning (Rename/organize data, columns & dropping Null/insignificant info)
Tech_Diversity = Tech_Data.rename(columns={"Employee Diversity in Tech": "Year",
                              "Unnamed: 1":" Company Type",
                              "Unnamed: 2": "Company Name",
                              "Unnamed: 3": "Female %",
                              "Unnamed: 4": "Male %",
                              "Unnamed: 5": "White %",
                              "Unnamed: 6": "Asian %",
                              "Unnamed: 7": "Latino %",
                              "Unnamed: 8": "Black %",
                               "Unnamed: 9": "Multi %",
                              "Unnamed: 10": "Other %",
                              "Unnamed: 11": "Undeclared %"})
Tech_Diversity.drop(Tech_Diversity.columns[[11]], axis=1, inplace=True)

Tech_Diversity = Tech_Diversity.drop(Tech_Diversity.index[[0]])

Tech_Diversity.head()
```

```python
Tech_Companies = Tech_Diversity["Company Name"].nunique()
Tech_Companies
```

```python
Tech_Diversity.dtypes
```

### Female Employees Per Company by Year with Total Average Percents

```python
Female = Tech_Diversity[['Company Name', 'Female %', 'Year']]
Female_Pivot = Female.pivot(index='Company Name',
                columns='Year',
               values ='Female %').apply(pd.to_numeric)

# Female_Pivot.head()
# Female_Pivot.dtypes

Female_Company_Averages = Female_Pivot.assign(Female_Employee_Avg = Female_Pivot.mean(axis=1)).round(1)
Female_Company_Averages.head()
```

### Male Employees Per Company by Year with Total Average Percents
